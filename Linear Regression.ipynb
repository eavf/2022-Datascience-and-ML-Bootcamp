```python
import pandas as pd
from pandas import DataFrame as df
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression as lr
```

```python
data = pd.read_csv('.\\static\\cost_revenue_clean.csv')
data
```

```python
data.describe()
```

```python
X = df(data,columns=['production_budget_usd'])
y = df(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(12,6))
plt.scatter(X,y, alpha = 0.3)
plt.title('Film cost vs. Global Revenue')
plt.xlabel('Production budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,430000000)
plt.show()
```

```python
regres = lr()
regres.fit(X, y)
```

Slope coeficient:

```python
# slope coeficient (sklon) - theta_1
regres.coef_
```

```python
#Intercept
regres.intercept_
```

Teraz aj s funkciou lineárnej regresie

```python
plt.figure(figsize=(12,6))
plt.scatter(X,y, alpha = 0.3)
plt.plot(X, regres.predict(X), color='red', linewidth=4)                       # nakreslí čiaru na grafe


plt.title('Film cost vs. Global Revenue')
plt.xlabel('Production budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,430000000)
plt.show()
```

Goodness of Fit
(hovorí o tom aká je variácia v zdroji dát 54 %)
Model vie vysvetliť 54 % variácií v reálnychg číslach - zdroja dát

```python
regres.score(X,y)
```
