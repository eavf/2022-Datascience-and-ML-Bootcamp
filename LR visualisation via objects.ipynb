```python
import pandas as pd
from pandas import DataFrame as df
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression as lr
```

```python
data = pd.read_csv('.\\static\\lsd_math_score_data.csv')
print (data)

time = data[['Time_Delay_in_Minutes']]              # musia byť zdvojené, lebo ak nebudú, vráti series, nie dataframe
lsd = data[['LSD_ppm']]
print(lsd)
score = data[['Avg_Math_Test_Score']]
```

```python
%matplotlib inline


plt.title('Tissue concentration of LSD over time', fontsize=17)

plt.xlabel('Time in minutes', fontsize=14)
plt.ylabel('Tissue LSD ppm', fontsize=14)
plt.text(x = 0, y = -0.2, s = 'Vovo and al. (1971)', fontsize=12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.ylim(1,7)
plt.xlim(0,500)

plt.style.use('classic')

plt.plot (time, lsd, color='g', linewidth=3)
plt.show()
```

```python
regr = lr()
regr.fit(lsd, score)
regr.coef_
```

```python
regr.coef_[0]
```

```python
regr.coef_[0][0]
```

```python
print("Theta1 : ", regr.coef_[0][0])
```

```python
print ("Theta 2 : ", regr.intercept_[0])
```

```python
print("Theta1 : ", regr.coef_[0][0])
print ("Intercept : (Theta 2) : ", regr.intercept_[0])
```

```python
regr.score(lsd, score)
```

```python
print("Theta1 : ", regr.coef_[0][0])
print ("Intercept : (Theta 2) : ", regr.intercept_[0])
print ("R-square : ", regr.score(lsd, score))

predicted_score = regr.predict(lsd)
```

```python
plt.scatter(lsd, score, color = 'blue', s = 100, alpha = 0.7)

plt.title('Arithmetic vs. LSD-25', fontsize=17)

plt.ylabel('Performance score', fontsize=14)
plt.xlabel('Tissue LSD ppm', fontsize=14)
plt.text(x = 0, y = -0.2, s = 'Vovo and al. (1971)', fontsize=12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlim(1,6.5)
plt.ylim(25,85)

plt.plot(lsd, predicted_score, color = 'red', linewidth=3)
plt.style.use('fivethirtyeight')







plt.show()
```
